```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from pylab import rcParams
```

```python

%matplotlib inline
rcParams['figure.figsize'] = 14, 7
```

```python
df_1 = pd.read_csv('../input/sample-data/sample_data_1.csv')
```

```python

df_1.head()
```

```python
plt.scatter(df_1.Pass_Fail, df_1.Score)
```

```python
dt_reg = DecisionTreeRegressor()
```

```python

x = df_1.Pass_Fail.values.reshape(-1,1)
y = df_1.Score.values
```

```python

dt_reg.fit(x,y)
```

```python

x_grid = np.arange(min(x), max(x), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(x, y, color="red")
plt.plot(x_grid, dt_reg.predict(x_grid), color="blue")
plt.show()
```

```python
dt_reg.predict(np.array([1]).reshape(-1, 1))
```
